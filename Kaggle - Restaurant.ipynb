```python
#Drop id
Train_data = train_df.drop(['Id'],axis=1)

#Change object to category
for c in Train_data.columns:
    if Train_data[c].dtype == object:
        Train_data[c] = Train_data[c].astype('category')
```

```python
from sklearn import feature_extraction
from __future__ import print_function

#Which variables are categorical
Categoric_features = [c for c in train_data if train_data[c].dtype.kind in 'O']
Train_Categoric_features = train_data[Categoric_features].to_dict(outtype='records')

# Construct a DictVectorizer to transform our dictionary to
# a binary on-hot encoded array for each row
vec = feature_extraction.DictVectorizer()
cat_vector = vec.fit_transform(Train_Categoric_features).toarray()
```

```python
# Construct a separate dataframe with the one-hot encoded data
# and set the column names by calling get_feature_names
df_vector = pd.DataFrame(cat_vector)
vector_columns = vec.get_feature_names()

# Drop the categorical columns and join the new one-hot 
# dataframe with the original dataframe
df_vector.columns = vector_columns
df_vector.index = train_data.index

df = train_data.drop(Categoric_features, axis=1)
df = df.join(df_vector)
```

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split

#Modeling
X_train_data = df.drop(['revenue'],axis=1)
Y_train_data = df['revenue']

#Split train and test
X_train, X_test, y_train, y_test = train_test_split(X_train_data, Y_train_data, test_size=0.2, random_state=2434)
```

```python
def rmse(x, y, coefs):
    yfit = np.polyval(coefs, x)
    return np.sqrt(np.mean((y - yfit) ** 2))
```
